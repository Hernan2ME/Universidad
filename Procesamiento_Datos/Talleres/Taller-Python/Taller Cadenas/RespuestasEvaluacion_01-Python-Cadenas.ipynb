***7.*** Convierte en mayúsculas el contenido de la variable **F**:

F = "You are wrong"

```python
### Asignamos el contenido a la cadena F
F = "You are wrong"
### Declaramos una nueva variable K que contenga a F pero en mayúsculas y usamos la función upper()
K = F.upper()
### Imprimimos los contenidos antes y después de haber sido convertidos en mayúsculas
print("Contenidos antes:", F, "\nContenidos después:", K)
```

***9.*** En la variable **G**, reemplaza la primera sub-cadena **Mary** con **Bob**:

```python
### Hacemos uso de la función replace() para reemplazar una sub-cadena con otra asignandola a una variable nueva para poder imprimirla
R = G.replace("Mary", "Bob")
### Imprimimos los contenidos antes y después de el reemplazo 
print("Contenidos antes del reemplazo: ", G, "\n\nContenidos después del reemplazo:", R)
```
